```python
import ast

def calculate_average_test_accuracy(file_path):
    try:
        with open(file_path, 'r') as file:
            total_accuracy = 0
            count = 0
            for line in file:
                # Convert the string representation of dictionary into an actual dictionary
                data = ast.literal_eval(line.strip())
                # Extract the test_accuracy_mean and accumulate
                total_accuracy += data['test_accuracy_mean']
                count += 1
            if count == 0:
                return 0  # Avoid division by zero, return 0 if no valid entries are found
            return total_accuracy / count, count
    except FileNotFoundError:
        print("The file was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

path_file ='/home/carol/NaroNet-main/Datasets/Results/1_EXP1/NaroNet/POLE Mutation_Copy number variation_MSI Status_Tumour Type/Cross_validation_results/folds_information.txt'
print(calculate_average_test_accuracy(path_file))
```

## Checking "Prediction values Fold i" file (Softmax)

```python
def calculate_accuracy(file_path):
    # Load the .xlsx file
    data = pd.read_excel(file_path)

    # Extract the third column ("Prediction") and convert string representations of lists to actual lists
    predictions = data['Prediction'].apply(lambda x: np.fromstring(x.strip('[]'), sep=' '))

    # Determine the prediction based on the values in the lists
    predicted_labels = predictions.apply(lambda x: 0 if x[0] > x[1] else 1)

    # Extract the true labels from the fourth column
    true_labels = data.iloc[:, 3]  # Assuming the fourth column is at index 3

    # Calculate accuracy
    accuracy = np.mean(predicted_labels == true_labels)
    return accuracy

file_path = "/home/carol/NaroNet-main/NaroNet-main/EXP2/NaroNet/Response/Cross_validation_results/Prediction_values_Response_Fold13.xlsx"
accuracy = calculate_accuracy(file_path)
print(f"The accuracy of the predictions is: {accuracy}")
```

```python
def compare_predictions(true_labels_path, predictions_path):
    # Load the true labels Excel file
    true_labels = pd.read_excel(true_labels_path)
    # Rename 'Image_Names' to 'Filename' and remove file extension for easier comparison
    true_labels['Filename'] = true_labels['Image_Names'].str.replace(r'\..*$', '', regex=True)

    # Load the predictions Excel file
    predictions = pd.read_excel(predictions_path)
    # Rename 'Subject_name' to 'Filename' for consistency
    predictions['Filename'] = predictions['Subject_name']

    # Convert numeric predictions to text labels (assuming the predicted label is in the fourth column)
    predictions['Copy number variation'] = predictions.iloc[:, 3].map({1: 'Low', 0: 'High'})

    # Merge the datasets on the Filename column
    merged_data = pd.merge(true_labels, predictions, on='Filename', suffixes=('_true', '_pred'))

    # Find differences: we'll focus on comparing a specific true label column (let's assume it's the second column after filename) with the prediction
    # Let's also assume that the true labels are in text format ('Low', 'High') directly.
    differences = merged_data[merged_data.iloc[:, 1] != merged_data['Copy number variation']]

    # Output the differences
    return differences
```

```python
def get_map(target_col):
    if target_col == 'Copy number variation':
        return {'Low':1, 'High':0}
    if target_col == 'MSI Status':
        return {'Stable':1, 'Instable':0}
    if target_col == 'POLE Mutation':
        return {'WT':1, 'Mutated':0}
    if target_col == 'Tumour Type':
        return {'Serous-like EC':1, 'Endometria carcinoma ':0}
```

```python
def fc(pred_path, true_path, target_col):
    predictions = pd.read_excel(pred_path)
    # Rename 'Subject_name' to 'Filename' for consistency
    predictions['Filename'] = predictions['Subject_name']
    predictions = predictions.drop('Subject_name', axis=1)

    true_labels = pd.read_excel(true_path)
    
    # Rename 'Image_Names' to 'Filename' and remove file extension for easier comparison
    true_labels['Filename'] = true_labels['Image_Names'].str.replace(r'\..*$', '', regex=True)
    true_labels = true_labels.drop('Image_Names', axis=1)
    true_labels['Label_true'] = true_labels[target_col].map(get_map(target_col))

    result = pd.merge(predictions, true_labels, on='Filename', how='inner')
    mismatched = result[result['Label'] != result['Label_true']]
    print(target_col)
    print(mismatched['Filename'])
```

```python
def replace_first_five(s, num):
    # Find the first occurrence of '5' in the string
    index = s.find('5')
    
    # If '5' is found, replace it with `num`
    if index != -1:
        s = s[:index] + str(num) + s[index+1:]
    
    return s

# Example usage
result = replace_first_five("/home/carol/NaroNet-main/Datasets/Results/5_EXP1/NaroNet/POLE Mutation_Copy number variation_MSI Status_Tumour Type", 3)
print(result)
```
